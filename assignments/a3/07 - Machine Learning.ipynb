## Modeling

A model is a specification of a mathematical (or probabilistic) relationship that exists between different variables. There are a wide range of models. Some models are just simple mathematical relationships between two or more variables. Some models are based on probability theory. We call them, probabilistic models.

```python
# first-order model
def fo_height(weight):
    return weight+100

# second-order model
def so_height(weight):
    return ((weight/24.0)**0.5)*100

# probabilistic model
def prob_height(weight):
    mu = ((weight/24.0)**0.5)*100 
    sigma = 10
    x = np.linspace(mu-4*sigma, mu+4*sigma, 100)
    plt.plot(x,norm.pdf((x-mu)/sigma),'-r')
```

## Overfitting and Underfitting

- Overfitting -- producing a model that performs well on the data we train it on but generalizes poorly to new data
- Underfitting -- producing a model that doesn't perform well even on the training data

```python
np.random.seed(101)
x = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
y = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
plt.plot(x,y,'sr')
p0 = np.poly1d(np.polyfit(x,y,0))
p1 = np.poly1d(np.polyfit(x,y,1))
p9 = np.poly1d(np.polyfit(x,y,9))
nx = np.linspace(-1,11,100)
plt.plot(nx,p1(nx),'-b',label="degree-1")
plt.plot(nx,p0(nx),'-m',label="degree-0")
plt.plot(nx,p9(nx),'-k',label="degree-9")
plt.ylim([-2,12])
plt.legend(loc=2)
```

## How to deal with Overfitting

How do we know if our model is overfitting? A fundamental way to check is to use different set of data for training the model and for testing the model.

It is common to split our data set into train and test set. For example, we use about 75% for training the model and 25% for testing. If the model was overfit to the training data, then it will perform poorly on the test data. On the other hand, if the model performs well on the test data, then we gain some confident that our model generalizes quite well.

Sometimes, you might want to train multiple models and then choose the "best" performing one. The common practice is to have another set of data called *validation set* for this purpose. To summarize:
 - training set -- for training models
 - validation set -- for choosing model and/or tuning params
 - test set -- for evaluating the final model

## Correctness of Models

How do we know if a preditive model is performing well or not? For each prediction, there are total of 4 outcomes:
 - True positive
 - False positive (Type 1 Error)
 - False negative (Type 2 Error)
 - True negative
 
**Example**

|             | spam           | not spam       |
|-------------|----------------|----------------|
| really spam | True positive  | False negative |
| not spam    | False positive | True nagetive  |

Now let's compute: accuracy, precision, recall, and f1_score

$ Accuracy = \frac{\text{# of correct prediction}}{\text{total # of prediction}}
= \frac{50+950000}{50+5000+15000+950000}
= 0.979 $

$ prediction = \frac{\text{# of True positive}}{\text{Total # of positive}} = \frac{50}{15050} = 0.003 $

$ recall = \frac{\text{# of True positive}}{\text{# of True positive + # of False negative}} = \frac{50}{5050} = 0.01 $

$ F_1 score = \text{The harmonic mean of precision and recall} = \frac{2 \times p \times r}{(p + r)} $

$ H = \frac{2}{\frac{1}{r_1} + \frac{1}{r_2}} = \frac{2 \times r_1 \times r_2}{r_1 + r_2} $

## Model Description

I select `Elongation` feature and use maximum and minimum value from the feature
to be the threshold for classifying classes

```python
# species-36 detector

# return separate train and test data
# 80% is train data
# 20% is test data
def separate_data(filename, delimiter=',', train_size=0.8):
    np.random.seed(999)
    data = np.genfromtxt(filename, delimiter=delimiter)
    msk = np.random.rand(len(data)) < train_size
    train = np.array(data[msk])
    test = np.array(data[~msk])
    return train, test

# http://stackoverflow.com/questions/24147278/how-do-i-create-test-and-train-samples-from-one-dataframe-with-pandas
```

```python
train_data, test_data = separate_data('leaf.csv')
```

```python
geranium_sp_filter = train_data[:,0] == 36
geranium_sp = train_data[geranium_sp_filter]
other_classes = train_data[~geranium_sp_filter]
```

```python
# use `Elongation` feature

feature_index = 4
feature = geranium_sp[:,feature_index]

max_gsp = np.max(feature)
min_gsp = np.min(feature)
```

```python
min_gsp, max_gsp
```

```python
def report_performance(data):
    true_positive = 0
    true_negative = 0
    false_positive = 0
    false_negative = 0
    class_index = 0
    n_data = len(data)
    selected_class = 36
    
    # classifying whether
    # the leaf is class 36
    for record in test_data:
        if min_gsp < record[feature_index] < max_gsp:
            if record[class_index] == selected_class:
                true_positive += 1
            else:
                false_negative += 1
        else:
            if record[class_index] == selected_class:
                false_positive += 1
            else:
                true_negative += 1
    accuracy = (true_positive + true_negative)/float(n_data)
    prediction = true_positive/float(true_positive+false_positive)
    recall = true_positive/float(true_positive+false_negative)
    f1_score = (2*prediction*recall)/(prediction+recall)
    print "Accuracy: {}".format(accuracy)
    print "Prediction: {}".format(prediction)
    print "recall: {}".format(recall)
    print "f1_score: {}".format(f1_score)
    print "-"*30
```

```python
print "Train data"
report_performance(train_data)
```

```python
print "Test data"
report_performance(test_data)
```
